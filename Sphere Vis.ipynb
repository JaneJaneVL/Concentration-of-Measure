### Poincare lemma. Experiments

Numerical Experiments demontstrating visualization 
of uniform distrubiton on a unit sphere in $\mathbb{R}^n$.

You may read about Poincare lemma, for example, in
*M. Lifshits Lectures on Gaussian Processes. Springer, 2012 (p. 40)*.

See also:
http://mathworld.wolfram.com/SpherePointPicking.html


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generating uniformly distributed numbers on a sphere
def get_rand_vec(dims):
    e = np.random.standard_normal(int(dims))
    return e / np.linalg.norm(e)
```

```python
#****************************************************
# number of points to generate
N = 1000
x = []
y = []
z = []
for i in range(1, N+1):
    e = get_rand_vec(3)
    x.append(e[0])
    y.append(e[1])
    z.append(e[2])
```

```python
# plots distribution of points
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
ax.plot(x,y,z,'bo')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
